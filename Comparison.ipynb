```python
import matplotlib.pyplot as plt
import pickle
```

```python
infile = open('history_bert','rb')
history_bert = pickle.load(infile)
infile.close()
```

```python
history_bert
```

```python
infile = open('history_lstm','rb')
history_lstm = pickle.load(infile)
infile.close()
```

```python
history_lstm
```

```python
infile = open('history_gru','rb')
history_gru = pickle.load(infile)
infile.close()
```

```python
history_gru
```

```python
plt.figure()
plt.xlabel('epochs')
plt.ylabel('Accuracy')
plt.plot(range(1,len(history_bert['loss'])+1), history_bert['accuracy'], label='BERT training accuracy')
plt.plot(range(1,len(history_lstm['loss'])+1), history_lstm['accuracy'], label='LSTM training accuracy')
plt.plot(range(1,len(history_gru['loss'])+1), history_gru['accuracy'], label='GRU training accuracy')
plt.scatter([8], [0.9169666666666667], marker='*', label='BERT test accuracy')
plt.scatter([6], [0.88170626], marker='*', label='GRU test accuracy', c='limegreen')
plt.scatter([5.9], [0.88100698], marker='*', label='LSTM test accuracy', c='orange')
plt.legend()
plt.savefig('accuracy_comparison')
plt.show()
```

```python
plt.figure()
plt.xlabel('epochs')
plt.ylabel('loss')
plt.plot(range(1,len(history_bert['loss'])+1), history_bert['loss'], label='BERT training loss')
plt.plot(range(1,len(history_lstm['loss'])+1), history_lstm['loss'], label='LSTM training loss')
plt.plot(range(1,len(history_gru['loss'])+1), history_gru['loss'], label='GRU training loss')
plt.legend()
plt.savefig('loss_comparison')
plt.show()
```
